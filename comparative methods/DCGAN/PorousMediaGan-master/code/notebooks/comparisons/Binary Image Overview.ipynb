```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import tifffile
%matplotlib inline
```

```python
names = ["beadpack", "berea", "ketton"]#, "Beadpack",
         #"Estaillades", "Bentheimer", "Doddington"]
direcs = ["../data/individual_images/beadpack/original/tiff/beadpack.tif",
          "../data/individual_images/berea/original/raw/berea.tif",
          "../data/individual_images/ketton/original/raw/ketton_orig.tif"]

images = []
for name in direcs:
    im = tifffile.imread(name)
    images.append(im[0, :,  :])
    print im.shape
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(48,16))

for ax, name, image, x_label, y_label in zip(axarr, names,
                                             images,
                                             [r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$', r'$x-Axis \ (Voxels)$'],
                                             [r'$y-Axis \ (Voxels)$', r'$y-Axis \ (Voxels)$', r'$y-Axis \ (Voxels)$']):

    ax.set_title(r'$'+name.title()+'$', size=45, y=1.02)


    ax.imshow(image, cmap='Greys',  interpolation='nearest')
    for item in ax.get_xticklabels():
        item.set_fontsize(28)
    for item in ax.get_yticklabels():
        item.set_fontsize(28)

    ax.set_xlabel(x_label, size=40)
    ax.set_ylabel(y_label, size=40)
    
for ax, lim, dim in zip(axarr, [501, 401, 257], [128, 64, 64]):
    ax.set_xlim(0, lim)
    ax.set_ylim(0, lim)
    
    ax.add_patch(patches.Rectangle((0+25, 25), dim, dim, fill=False, hatch='\\', linestyle='solid', linewidth=15, color="red"))
        
    
fig.canvas.draw()

plt.show()
fig.savefig("figures/image_views.png", bbox_extra_artists=None, bbox_inches='tight', dpi=300)
plt.close()
```
