```python
import numpy as np
import tifffile
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sampled_images = []
for i in range(43, 64, 1):
    im_in = tifffile.imread("../../../data/ketton/sampled/postprocessed/tiff/200/ketton_desp_64_"+str(i)+".tiff")
    sampled_images.append(im_in)

original_images = []
for i in range(0, 7, 1):
    im_in = tifffile.imread("../../../data/ketton/original/postprocessed/tiff/200/ketton_flow_sub"+str(i)+".tiff")
    original_images.append(im_in)
```

```python
fig, ax = plt.subplots(5, 5, figsize=(10, 10))
#fig.subplots_adjust(hspace=0, wspace=0)
z = 100

for i in range(0, 5):
    img = ax[0, i].imshow(sampled_images[i][:, :, z], interpolation='nearest', cmap="Greys")

for i in range(5, 10):
    img = ax[1, i-5].imshow(sampled_images[i][:, :, z], interpolation='nearest', cmap="Greys")

for i in range(10, 15):
    img = ax[2, i-10].imshow(sampled_images[i][:, :, z], interpolation='nearest', cmap="Greys")

for i in range(15, 20):
    img = ax[3, i-15].imshow(sampled_images[i][:, :, z], interpolation='nearest', cmap="Greys")


for j in range(5):
    img = ax[4, j].imshow(original_images[j][:, :, z], interpolation='nearest', cmap="Greys")
  
for a in ax.flatten():
    a.set_xticklabels([])
    a.set_yticklabels([])
    a.set_aspect('equal')
    a.set_xticks([])
    a.set_yticks([])

ax[0, 0].set_ylabel(r"$Synthetic \ Image$")
ax[4, 0].set_ylabel(r"$Training \ Image$")

plt.subplots_adjust(wspace=0.05, hspace=0.05)

plt.axis('off')

fig.savefig("../../../paper/figures/ketton_comparison.png", bbox_inches='tight', dpi=300)
```
